```python
import numpy as np
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('DigMNIST.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
X = data.drop(columns=['label'])  
X.head()
```

```python
X_std = StandardScaler().fit_transform(X)
```

```python
pca = PCA().fit(X_std)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlim(0,400,1)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
```

```python
Xpca = PCA(n_components=400)
Ypca= Xpca.fit_transform(X_std)
```

```python
pca_df = pd.DataFrame(data = Ypca)
pca_df
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.show()
```

### Inverse PCA

```python
X_reconstructed = pca.inverse_transform(X_std)
inverse_pca_df = pd.DataFrame(data = X_reconstructed)
```
